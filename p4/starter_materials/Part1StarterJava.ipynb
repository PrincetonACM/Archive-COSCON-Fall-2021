```
// This is just used to run command line commands.
public static String execCmd(String cmd) throws java.io.IOException {
    java.util.Scanner s = new java.util.Scanner(Runtime.getRuntime().exec(cmd).getInputStream()).useDelimiter("\\A");
    return s.hasNext() ? s.next() : "";
}
```

```
// This function can be used to write every element in a list to a file (one element per line)
// You may find this useful to generate the output text file
// For example, to save the data in predictions_array_list to a text file titled YeastPart1.txt in 
// your Colab Notebooks folder, you can use the code below:
// writeFile(predictions_array_list, "YeastPart1.txt", "drive/MyDrive/Colab Notebooks/");

public static void writeFile(ArrayList<String> values, String filename, String directory){
  BufferedWriter wr = null;
  System.out.println(directory+filename);
  try {
      wr = new BufferedWriter(new FileWriter(directory+filename));
      for (String var : values) {
          wr.write(var);
          wr.newLine();
      }
      wr.close();

  } catch (Exception e) {
  }
}
```
